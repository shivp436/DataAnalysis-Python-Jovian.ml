```python
# a function to get a dictionary structure out of a file
import math

def calcEMI(values):
    amt = values[0]
    durn = values[1]
    rate = values[2]/12
    down_pmt = values[3]

    loan = amt-down_pmt
    try:
        emi = loan * rate * ((1+rate) ** durn) / (((1+rate)**durn)-1)
    except ZeroDivisionError:
        emi = loan / durn
    emi = math.ceil(emi)
    return emi

def getEMI(filename):
    lines = None
    with open(filename) as file:
        lines = file.readlines()
        headers = lines[0].strip().split(',')
        headers.append('emi')

        lines = lines[1:]
        line_no = len(lines)

        for line in range(line_no):
            curr_line = lines[line]
            curr_line_arr = curr_line.strip().split(',')
            values = []
            for val in curr_line_arr:
                if val == '':
                    values.append(0.0)
                else:
                    values.append(float(val))
            emi = calcEMI(values)
            values.append(emi)
            curr_line_dict = dict(zip(headers, values))
            lines[line] = curr_line_dict

    return lines
```

```python
def dictToMat(dic):
    mat = []

    # headers
    headers = []
    for key in dic[0].keys():
        headers.append(key)
    mat.append(headers)

    # values
    for line in dic:
        values = []
        for header in headers:
            values.append(line[header])
        mat.append(values)
    return mat
```

```python
def saveToCsv(mat, opFileName):
    headers = ','.join(mat[0])
    values = mat[1:]

    np.savetxt(
        opFileName
        , values
        , fmt = '%.2f'
        , delimiter = ' , '
        , header = headers
        , comments = ''
    )

    print("Matrix saved to ", opFileName)
```
